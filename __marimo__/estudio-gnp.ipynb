```
import marimo as mo
from timeseries import TimeSeries
import pandas as pd
```

```
df = pd.read_spss("./data/gnp.sav")


def parse_date(date: str) -> str:
    trim = {1: "01", 2: "04", 3: "07", 4: "10"}
    q, y = date.split()
    num = int(q[1])
    return f"01-{trim[num]}-{y}"


df["Tiempo"] = df["date_"].apply(parse_date)
print(df.head())
```

```
gnp = TimeSeries.from_dataframe(df, "indice", "Tiempo")
gnp.plot()
```

```
decomp = gnp.decompose(period=4)
decomp.plot()
```

```
gnp.plot_correlograms()
```

Vemos que la grfica de FAS decae muy lentamente, esto se corresponde con la tendencia que observamos en la grfica.

En la grfica de FAP vemos una correlacin casi total en el lag 1 y despus barras no significativas, parece un camino aleatorio.

```
gnp_est = gnp.log().diff(order=0, seasonal_lag=4, seasonal_order=1)
gnp_est.plot()
```

Al aplicar una transformacin logartmica y luego diferenciar estacionalmente una vez hemos eliminazo por completo la tencendia, por lo que es casi seguro que nuestro parmetro de diferenciacin sea d = 1.

```
gnp_est.plot_correlograms()
```

En el nuevo grfico FAS vemos un decaimiento ms rpido, aunque no abrupto, con barras significativas en los lags 1, 2 y ms o menos 3. Este decaimiento nos sugiere q = 0, es decir, sin componente MA.

En la grfica FAP vemos barras significativas en los primeros 3 lags. Sin embargo, la barra correspondiente al lag 3 es significativa por bastante poco, sobre todo en comparacin con los lags 1 y 2. Por ello prefiero como modelo candidato un AR(2).
